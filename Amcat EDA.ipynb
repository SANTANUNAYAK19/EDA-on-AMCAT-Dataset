### Creating Function for Univariate Analysis on categorical and numerical data

```python
def categorical_univariate_analysis(cat_data):
    for columns in cat_data:
        print("*"*10, columns, "*"*10)
        print(cat_data[columns].agg(['count', 'nunique', 'unique']))
        print('Value Counts: \n', cat_data[columns].value_counts())
        print()
```

```python
categorical_univariate_analysis(cat_df)
```

```python
def numerical_univariate_analysis(num_data):
    for columns in num_data:
        print("*"*10, columns, "*"*10)
        print(num_data[columns].agg(['min', 'max', 'mean', 'median', 'std']))
        print()
```

```python
numerical_univariate_analysis(num_df)
```

### Bivariate Analysis

```python
rel_ds=pd.DataFrame(df.groupby("Designation")["Specialization"].value_counts())
```

```python
new = rel_ds.nlargest(columns='count',n = 10)
new
```

```python
plt.figure(figsize=(3,3))
sns.barplot(data=new, x='count', y='Designation', hue='Specialization', dodge=False)
plt.legend(title='Specialization', bbox_to_anchor=(1.05, 1), loc='upper left') 
```

```python
import warnings
warnings.filterwarnings('ignore')

# Scatter plot for numerical relationships
sns.scatterplot(x='Salary', y='10percentage', data=df)
plt.title('Scatterplot of Salary vs 10th Grade Percentage')
plt.show()

# Pairplot for numerical relationships
sns.pairplot(df[['Salary', '10percentage', '12percentage', 'collegeGPA']])
plt.show()

# Swarmplot for categorical and numerical relationship
sns.swarmplot(x='JobCity', y='Salary', data=df)
plt.title('Swarmplot of Salary vs Job City')
plt.xticks(rotation=90)
plt.show()

# Stacked bar plot for categorical relationships
pd.crosstab(df['JobCity'], df['Gender']).plot(kind='bar', stacked=True)
plt.title('Stacked Bar Plot of Gender Distribution by Job City')
plt.show()
```

```python
SS_plot = pd.DataFrame(df.groupby("Specialization")["Salary"].max().nlargest(10))

plt.figure(figsize=(3,3))
sns.barplot(data=SS_plot,x="Salary",y="Specialization")
plt.xticks(rotation=90)
plt.show()
```
